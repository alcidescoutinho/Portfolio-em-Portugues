```python
def quartil(col):

  df_quant = df_final.loc[df_final['situacao_confirmada'] == col, 'idade']
  q1, q3 = np.quantile(df_quant,[0.25,0.75])
  q2 = df_quant.median()
  qr = q3-q1
  li = q1 - (1.5*qr)
  if li < 0:
    li = df_quant.min()
  ls = q3 + (1.5*qr)
  if ls > df_quant.max():
    ls=df_quant.max()
  return li,q1, q2, q3, ls
```

```python
def grafico_3():
  # Comum
  figura, ax = plt.subplots(1,1,figsize=(35,15))

  # Figura 1
  fig = sns.boxplot(ax=ax, data=df_final, x='idade',y='situacao_confirmada',palette='crest')
  plt.title('Quartil das situações dos pacientes', size=35, y=1.05)
  for i,z in zip(range(5), df_final['situacao_confirmada'].unique()):
    li,q1, q2, q3, ls = quartil(z)
    interquartis_1 = [q1, q2, q3]
    limits = [li, ls]
    for ni in interquartis_1:
      ax.annotate(int(ni),xy=(ni,i+0.4),size=15,color='black',textcoords='data',xytext=(ni, i+0.5),
                    arrowprops=dict(facecolor='black', shrink=0.05),
                    va='top', ha='center')
    for ni in limits:
      ax.annotate(int(ni),xy=(ni,i+0.3),size=15,color='black')
  plt.ylabel('Situações dos pacientes', size=25)
  plt.xlabel('Idade', size=25)
  plt.yticks(size=15)

  salvar_figura('quartil',figura)
grafico_3()
```

```python
def quartis(df):
  q1,q2, q3 = np.quantile(df,[0.25,0.5,0.75])
  qr = q3-q1
  li = q1 - (1.5*qr)

  if li < 0:
    li = df.min()
  ls = q3 + (1.5*qr)
  if ls > df.max():
    ls=df.max()
  return li,q1, q2, q3, ls
```
